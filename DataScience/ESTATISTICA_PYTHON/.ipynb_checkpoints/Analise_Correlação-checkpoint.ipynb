```python
# Importar bibliotecas gerais
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importar bibliotecas para análises
from scipy import stats as st
import pingouin as pg

# Carregar dados
dados_tratamentos = pd.read_csv('dados_tratamentos.csv', sep = ';')
dados_nps = pd.read_csv('nps_example.csv', sep = ';')
dados_bolsa = pd.read_csv('dados_bolsa.csv', sep = ';', decimal = ',')
```

```python
import pandas as pd

# Criando a tabela cruzada
tabela = pd.crosstab(dados_tratamentos['tratamento'], dados_tratamentos['evolucao'])

# Para melhorar a formatação e exibição da tabela, você pode usar:
# 1. Renomear as colunas (se necessário)
tabela.columns = ['Evolução ' + str(col) for col in tabela.columns]

# 2. Adicionar um título para o índice
tabela.index.name = 'Tratamento'

# Mostrar a tabela formatada
print(tabela)
```

```python
import numpy as np
import scipy.stats as st

# qui-quadrado da tabela de contingência
qui_quadrado = st.chi2_contingency(tabela)[0]

# tamanho da amostra
n = np.sum(np.sum(tabela))

# cálculo do coeficiente Phi
Phi = np.sqrt(qui_quadrado / n)
Phi

#0,21 pode ser considerado como associacao muito forte
```

```python
# obter valor do qui quadrado
qui, p, gl, t_esp = st.chi2_contingency(tabela)
# ver os resultados
qui, p

#Como o P value e menor que 0,05, devemos descartar a hipotese nula, consluindo assim que existe uma associacao entre o tratamento e a melhora dos paciente
```

```python
dados_nps_filtrados = dados_nps[
    (dados_nps['response_status'] == 'Complete') &
    (dados_nps['nps_score'].notna()) &
    (dados_nps['gender'].notna())
].reset_index()

dados_nps_filtrados.head(5)
```

```python
def create_nps_groups(x):
    if x <= 5:
        return 'detratores'
    elif x >= 9:
        return 'promotores'
    else:
        return 'neutros'

dados_nps_filtrados['nps_groups'] = dados_nps_filtrados['nps_score'].apply(lambda x: create_nps_groups(x))

dados_nps_filtrados.groupby('nps_groups').size().to_frame('n')
```

```python
tabela = pd.crosstab(dados_nps_filtrados['gender'], dados_nps_filtrados['nps_groups'])

tabela
```

```python
vcramer_1 = st.contingency.association(tabela, method='cramer')

vcramer_1

# Avaliação do resultado do coeficiente de associação de Cramér
if vcramer_1 < 0.1:
    print("Associação muito fraca entre as variáveis.")
elif 0.1 <= vcramer_1 < 0.3:
    print("Associação fraca entre as variáveis.")
elif 0.3 <= vcramer_1 < 0.5:
    print("Associação moderada entre as variáveis.")
elif 0.5 <= vcramer_1 < 0.7:
    print("Associação considerável entre as variáveis.")
elif 0.7 <= vcramer_1 < 1:
    print("Associação forte entre as variáveis.")
else:
    print("Valor do coeficiente de associação inválido.")
```

```python
# Valores para o cálculo do coeficiente de associação de Cramér usando o segundo método
qui_quadrado = st.chi2_contingency(tabela)[0]
n = np.sum(np.sum(tabela))
minshape = min(tabela.shape) - 1

# Cálculo do coeficiente de associação de Cramér (segundo método)
vcramer_2 = np.sqrt(qui_quadrado / (n * minshape))

# Avaliação do resultado do coeficiente de associação de Cramér (segundo método)
if vcramer_2 < 0.1:
    print("Associação muito fraca entre as variáveis.")
elif 0.1 <= vcramer_2 < 0.3:
    print("Associação fraca entre as variáveis.")
elif 0.3 <= vcramer_2 < 0.5:
    print("Associação moderada entre as variáveis.")
elif 0.5 <= vcramer_2 < 0.7:
    print("Associação considerável entre as variáveis.")
elif 0.7 <= vcramer_2 < 1:
    print("Associação forte entre as variáveis.")
else:
    print("Valor do coeficiente de associação inválido.")
```

```python
# Teste Qui-quadrado
qui, p, gl, t_esp = st.chi2_contingency(tabela)
qui, p
```

Qui-quadrado (qui): 28.87689583154292
Valor-p (p): 5.363666207418583e-07
Esses resultados do teste Qui-quadrado são significativos para analisar a associação entre as variáveis presentes na tabela de contingência. Aqui está o significado desses resultados:

Qui-quadrado (qui): Esse valor representa a estatística de teste do Qui-quadrado. Quanto maior for o valor do Qui-quadrado, mais evidências existem contra a hipótese nula de independência entre as variáveis. Neste caso, o valor é 28.88, indicando uma diferença significativa entre as frequências observadas e as esperadas, o que sugere uma associação entre as variáveis.

Valor-p (p): O valor-p é a probabilidade de observar uma estatística de teste igual ou mais extrema do que aquela observada, sob a suposição de que a hipótese nula (geralmente a hipótese de independência entre as variáveis) é verdadeira. No caso apresentado, o valor-p é 5.36e-07 (ou 0.000000536). Um valor-p pequeno (geralmente considerado menor que 0.05) sugere fortemente rejeitar a hipótese nula, indicando que as variáveis não são independentes.

Em resumo, o Qui-quadrado relativamente alto e o valor-p muito baixo indicam fortes evidências contra a hipótese de independência entre as variáveis, sugerindo que há uma associação significativa entre elas.

```python
df = pd.read_csv('dados_credito_small.csv')

df.head()
```

```python
df[df.isnull()]

df= df.dropna()
```

```python
tabela_credito = pd.crosstab(df['default'], df['renda_cat'])

tabela_credito
```

```python
vcramer_1 = st.contingency.association(tabela_credito, method='cramer')

vcramer_1

# Avaliação do resultado do coeficiente de associação de Cramér
if vcramer_1 < 0.1:
    print("Associação muito fraca entre as variáveis.")
elif 0.1 <= vcramer_1 < 0.3:
    print("Associação fraca entre as variáveis.")
elif 0.3 <= vcramer_1 < 0.5:
    print("Associação moderada entre as variáveis.")
elif 0.5 <= vcramer_1 < 0.7:
    print("Associação considerável entre as variáveis.")
elif 0.7 <= vcramer_1 < 1:
    print("Associação forte entre as variáveis.")
else:
    print("Valor do coeficiente de associação inválido.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Gráfico de dispersão usando Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(df['renda'], df['divida'], c='blue', label='Renda vs Dívida')
plt.scatter(df['renda'], df['outras_div'], c='red', label='Renda vs Outras Dívidas')
plt.xlabel('Renda')
plt.ylabel('Valores')
plt.title('Gráfico de Dispersão: Renda vs Dívida e Outras Dívidas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Selecionando apenas as colunas 'renda', 'divida' e 'outras_div' para calcular a correlação
colunas_selecionadas = df[['renda', 'divida', 'outras_div']]

# Calculando a correlação utilizando o método 'corr()'
correlacao = colunas_selecionadas.corr(method='pearson')

# Imprimindo a matriz de correlação
print(correlacao)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor - Matriz de Correlação')
plt.show()
```
