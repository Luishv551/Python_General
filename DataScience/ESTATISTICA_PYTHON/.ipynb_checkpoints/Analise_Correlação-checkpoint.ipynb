```python
import pandas as pd

# Criando a tabela cruzada
tabela = pd.crosstab(dados_tratamentos['tratamento'], dados_tratamentos['evolucao'])

# Para melhorar a formatao e exibio da tabela, voc pode usar:
# 1. Renomear as colunas (se necessrio)
tabela.columns = ['Evoluo ' + str(col) for col in tabela.columns]

# 2. Adicionar um ttulo para o ndice
tabela.index.name = 'Tratamento'

# Mostrar a tabela formatada
print(tabela)
```

```python
import numpy as np
import scipy.stats as st

# qui-quadrado da tabela de contingncia
qui_quadrado = st.chi2_contingency(tabela)[0]

# tamanho da amostra
n = np.sum(np.sum(tabela))

# clculo do coeficiente Phi
Phi = np.sqrt(qui_quadrado / n)
Phi

#0,21 pode ser considerado como associacao muito forte
```

```python
#obtervalordoquiquadrado
qui, p, gl, t_esp = st.chi2_contingency(tabela)
#ver os resultados
qui, p

#Como o P value e menor que 0,05, devemos descartar a hipotese nula, consluindo assim que existe uma associacao entre o tratamento e a melhora dos paciente
```

```python
dados_nps_filtrados = dados_nps[
    (dados_nps['response_status'] == 'Complete') &
    (dados_nps['nps_score'].notna()) &
    (dados_nps['gender'].notna())
].reset_index()

dados_nps_filtrados.head(5)
```

```python
def create_nps_groups(x):
    if x <= 5:
        return 'detratores'
    elif x >= 9:
        return 'promotores'
    else:
        return 'neutros'

dados_nps_filtrados['nps_groups'] = dados_nps_filtrados['nps_score'].apply(lambda x: create_nps_groups(x))

dados_nps_filtrados.groupby('nps_groups').size().to_frame('n')
```

```python
tabela = pd.crosstab(dados_nps_filtrados['gender'], dados_nps_filtrados['nps_groups'])

tabela
```

```python
vcramer_1 = st.contingency.association(tabela, method='cramer')

vcramer_1

# Avaliao do resultado do coeficiente de associao de Cramr
if vcramer_1 < 0.1:
    print("Associao muito fraca entre as variveis.")
elif 0.1 <= vcramer_1 < 0.3:
    print("Associao fraca entre as variveis.")
elif 0.3 <= vcramer_1 < 0.5:
    print("Associao moderada entre as variveis.")
elif 0.5 <= vcramer_1 < 0.7:
    print("Associao considervel entre as variveis.")
elif 0.7 <= vcramer_1 < 1:
    print("Associao forte entre as variveis.")
else:
    print("Valor do coeficiente de associao invlido.")
```

```python
# Valores para o clculo do coeficiente de associao de Cramr usando o segundo mtodo
qui_quadrado = st.chi2_contingency(tabela)[0]
n = np.sum(np.sum(tabela))
minshape = min(tabela.shape) - 1

# Clculo do coeficiente de associao de Cramr (segundo mtodo)
vcramer_2 = np.sqrt(qui_quadrado / (n * minshape))

# Avaliao do resultado do coeficiente de associao de Cramr (segundo mtodo)
if vcramer_2 < 0.1:
    print("Associao muito fraca entre as variveis.")
elif 0.1 <= vcramer_2 < 0.3:
    print("Associao fraca entre as variveis.")
elif 0.3 <= vcramer_2 < 0.5:
    print("Associao moderada entre as variveis.")
elif 0.5 <= vcramer_2 < 0.7:
    print("Associao considervel entre as variveis.")
elif 0.7 <= vcramer_2 < 1:
    print("Associao forte entre as variveis.")
else:
    print("Valor do coeficiente de associao invlido.")
```

```python
# Teste Qui-quadrado
qui, p, gl, t_esp = st.chi2_contingency(tabela)
qui, p
```

Qui-quadrado (qui): 28.87689583154292
Valor-p (p): 5.363666207418583e-07
Esses resultados do teste Qui-quadrado so significativos para analisar a associao entre as variveis presentes na tabela de contingncia. Aqui est o significado desses resultados:

Qui-quadrado (qui): Esse valor representa a estatstica de teste do Qui-quadrado. Quanto maior for o valor do Qui-quadrado, mais evidncias existem contra a hiptese nula de independncia entre as variveis. Neste caso, o valor  28.88, indicando uma diferena significativa entre as frequncias observadas e as esperadas, o que sugere uma associao entre as variveis.

Valor-p (p): O valor-p  a probabilidade de observar uma estatstica de teste igual ou mais extrema do que aquela observada, sob a suposio de que a hiptese nula (geralmente a hiptese de independncia entre as variveis)  verdadeira. No caso apresentado, o valor-p  5.36e-07 (ou 0.000000536). Um valor-p pequeno (geralmente considerado menor que 0.05) sugere fortemente rejeitar a hiptese nula, indicando que as variveis no so independentes.

Em resumo, o Qui-quadrado relativamente alto e o valor-p muito baixo indicam fortes evidncias contra a hiptese de independncia entre as variveis, sugerindo que h uma associao significativa entre elas.

```python
df = pd.read_csv('dados_credito_small.csv')

df.head()
```

```python
tabela_credito = pd.crosstab(df['default'], df['renda_cat'])

tabela_credito
```

```python
vcramer_1 = st.contingency.association(tabela_credito, method='cramer')

vcramer_1

# Avaliao do resultado do coeficiente de associao de Cramr
if vcramer_1 < 0.1:
    print("Associao muito fraca entre as variveis.")
elif 0.1 <= vcramer_1 < 0.3:
    print("Associao fraca entre as variveis.")
elif 0.3 <= vcramer_1 < 0.5:
    print("Associao moderada entre as variveis.")
elif 0.5 <= vcramer_1 < 0.7:
    print("Associao considervel entre as variveis.")
elif 0.7 <= vcramer_1 < 1:
    print("Associao forte entre as variveis.")
else:
    print("Valor do coeficiente de associao invlido.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Grfico de disperso usando Matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(df['renda'], df['divida'], c='blue', label='Renda vs Dvida')
plt.scatter(df['renda'], df['outras_div'], c='red', label='Renda vs Outras Dvidas')
plt.xlabel('Renda')
plt.ylabel('Valores')
plt.title('Grfico de Disperso: Renda vs Dvida e Outras Dvidas')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Selecionando apenas as colunas 'renda', 'divida' e 'outras_div' para calcular a correlao
colunas_selecionadas = df[['renda', 'divida', 'outras_div']]

# Calculando a correlao utilizando o mtodo 'corr()'
correlacao = colunas_selecionadas.corr(method='pearson')

# Imprimindo a matriz de correlao
print(correlacao)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title('Mapa de Calor - Matriz de Correlao')
plt.show()
```
