# NA Values

```python
import pandas as pd

# Ler o arquivo CSV ajustando os separadores
dados_hatco = pd.read_csv('hatco.csv', decimal=',')
```

```python
dados_hatco.head()
```

```python
missings_por_linha = pd.DataFrame({
    'id': dados_hatco['id'].tolist(),
    'n_missings': dados_hatco.isnull().sum(axis=1).tolist()
})

n_colunas = dados_hatco.shape[1] - 1

# Calcular a porcentagem de valores ausentes por linha
missings_por_linha['perc_missings'] = missings_por_linha['n_missings'] / n_colunas

# Classificar as linhas com base na porcentagem de valores ausentes
top_10_linhas_com_mais_missings = missings_por_linha.sort_values('perc_missings', ascending=False).head(10)

print(top_10_linhas_com_mais_missings)
```

```python
pd.crosstab(dados_hatco.tamanho_firma.fillna('NA'), dados_hatco.area_industria.fillna('NA'))
```

```python
dados_hatco[['velocidade_entrega', 'qualidade_produto']].corr()
```

```python
dados_hatco[['velocidade_entrega', 'qualidade_produto']].fillna(0).corr()
```

```python
dados_hatco.dropna().shape
```

```python
dados_hatco['area_industria'].value_counts(dropna=False).to_frame('n').reset_index().rename(columns={'index': 'area_industria'})
```

```python
dados_hatco['area_industria'].fillna('outra_area').value_counts(dropna=False).to_frame('n').reset_index().rename(columns={'index': 'area_industria'})
```

```python
stats = dados_hatco[['flexibilidade_preco', 'imagem_fornecedor']].describe()

stats
```

```python
dados_hatco2 = dados_hatco \
    .assign(flexibilidade_preco2 = dados_hatco['flexibilidade_preco'].fillna(stats['flexibilidade_preco'].loc['50%'])) \
    .assign(imagem_fornecedor2 = dados_hatco['imagem_fornecedor'].fillna(stats['imagem_fornecedor'].loc['mean']))
dados_hatco2[['id', 'flexibilidade_preco', 'flexibilidade_preco2', 'imagem_fornecedor', 'imagem_fornecedor2']].tail(5)
```

# Outliers

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import scale
from sklearn.covariance import EmpiricalCovariance, MinCovDet

dados_credito = pd.read_csv('dados_credito_small.csv', sep=',', decimal='.')
```

```python
# dados credito
dados_credito.head()
```

```python
import numpy as np
from sklearn.preprocessing import scale

# Lista de números (simulando os dados)
numeros = np.array([15, 20, 30, 10, 25])

# Normalização dos dados para média 0 e desvio padrão 1
numeros_normalizados = scale(numeros)
print(numeros_normalizados)
```

```python
dados_credito['outras_div_z'] = scale(dados_credito['outras_div'])
dados_credito.sort_values('outras_div_z', ascending=False)
```

```python
dados_credito[['idade', 'outras_div']].boxplot()
```

```python
dados_credito.plot.scatter(x='divida', y='outras_div')

plt.show()
```

```python
from sklearn.covariance import MinCovDet

robust_cov = MinCovDet().fit(dados_credito[['idade', 'renda', 'divida']])
dados_credito['mahalanobis_d'] = robust_cov.mahalanobis(dados_credito[['idade', 'renda', 'divida']])
dados_credito.head()
```

```python
dados_credito.plot.scatter(x='id', y='mahalanobis_d')

plt.show()
```

```python
dados_credito['outras_div_cat'] = pd.qcut(dados_credito['outras_div'], q=4, labels=['baixa', 'baixa-media', 'media-alta', 'alta'])
dados_credito[['id', 'idade', 'outras_div', 'outras_div_cat']].head()
```

```python
wins_values = dados_credito['outras_div'].quantile([0.05, 0.9]).to_list()

wins_values
```

```python
dados_credito['outras_div_wins'] = dados_credito['outras_div'].clip(wins_values[0], wins_values[1])
```

```python
dados_credito[['id', 'outras_div', 'outras_div_wins']].head(10)
```

```python
# box plot
dados_credito[['outras_div', 'outras_div_wins']].boxplot()

plt.show()
```
